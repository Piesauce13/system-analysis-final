```python
numbers = [100,2,3,4,5]

def find_min(numbers):
    min = numbers[0]
    for i in numbers:
        if i < min:
            min = i
    return min

print("The mininmum number is:", find_min(numbers))

def find_max(numbers):
    max = numbers[0]
    for i in numbers:
        if i > max:
            max = i
    return max

print("The maximum number is:", find_max(numbers))


def find_sum(numbers):
    sum = 0
    for i in numbers:
        sum = sum + i
    return sum

print("The sum of the numbers is:", find_sum(numbers))

def find_avg(numbers):
    avg = 0
    avg = find_sum(numbers) / len(numbers)
    return avg

numbers = [100,2,3,4,5]

def search(numbers):
    number = int(input("Enter a number to search for:"))
    for idx, i in enumerate(numbers):
        if i == number:
            print(f"Number {number} is at index {idx}")
            return
    print("Number not found")
        
search(numbers)
```

Exercise 3

```python
def print_rect(w, h):
    for i  in range(h):
        print("*" * w)

print_rect(5, 3)

def print_hollow_rect(width, height, char="*"):
    for i in range(height):
        if i == 0 or i == height - 1:
            print(char * width)
        else:
            print(char + " " * (width - 2) + char)

print_hollow_rect(5, 4)   

def print_tri(h):
    for i in range(1, h):
        print("*" * i)

print_tri(8)
```

```python
movies = [
    {
        "title": "The Shawshank Redemption",
        "director": "Frank Darabont",
        "year": 1994,
        "genre": ["Drama"]
    },
    {
        "title": "The Godfather",
        "director": "Francis Ford Coppola",
        "year": 1972,
        "genre": ["Crime", "Drama"]
    },
]
```

```python
def add_movie():
    print("Please add info about the movie you want to add")

    title = input("Enter the movie title:")
    director = input("Enter the movie director:")
    year = int(input("Enter the movie year of release:"))
    genre = input("Ã‹nter the movie genre:")

    new_movie = {
        "title":title,
        "director":director,
        "year":year,
        "genre":genre
    }

    movies.append(new_movie)
    print(f"{title} has been added successfully")
    print(movies)

add_movie()
```

```python
def search_movie():
    search_title = input("Enter the title of the movie you wish to search:")

    for movie in movies:
        if movie["title"] == search_title:
            print(movie)
            return
    
    print("Movie not found")

search_movie()   
```

```python
def filter_genre():
    genre_search = input("Enter the genre you wish to filter:")

    for movie in movies:
        if movie["genre"] == genre_search:
            print(movie)
            return
    
    print("Movie not found")

filter_genre()   
```

```python
def filter_year():
    year_search = int(input("Enter the year you wish to filter:"))

    for movie in movies:
        if movie["year"] == year_search:
            print(movie)
            return
    
    print("Movie not found")

filter_year()   
```

```python
def display_movies():
    for idx, movie in enumerate(movies):
        print(f"""Movie #{idx+1}
              Title: {movie['title']}
              Director: {movie['director']}
              Year of Release: {movie['year']}
              Genre: {movie['genre']}""")
        
    return
    print("Movie not found")

display_movies()   
```
